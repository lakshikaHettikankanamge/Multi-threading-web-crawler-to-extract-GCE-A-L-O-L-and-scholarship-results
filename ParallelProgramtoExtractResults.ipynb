# mutli threading parallel program

```
from threading import Thread
import requests
import json
from time import perf_counter 
```

```
#function to extract A/l results - ur is any 7 digit number
def alevel(ur):
  
    base = "https://result.doenets.lk/result/service/AlResult?index="+str(ur)+"&nic="
    
    r = requests.get(base)
    y = json.loads(r.text)
    print(ur)
    
    if(y["examination"]=="G.C.E. (A/L) EXAMINATION (After Rescrutiny)"):
      print (r.text)
      collection1.insert_one( y)
```

```
#function to extract O/l results - ur is any 8 digit number
def olevel(ur):
  
    base = "https://result.doenets.lk/result/service/OlResult?index="+str(ur)+"&nic="
    
    r = requests.get(base)
    y = json.loads(r.text)
    print(ur)
    
    if(y["examination"]=="G.C.E. (O/L) EXAMINATION"):
      print (r.text)
      collection2.insert_one( y)
      #with open('gdrive/MyDrive/olevel.json') as json_file: 
        #data = json.load(json_file) 
        #data.append(y)
```

```
#function to extract scholarship results - ur 7 digits
def scholarship(ur):
  
    base = "https://result.doenets.lk/result/service/GvResult/"+str(ur)
    
    r = requests.get(base)
    y = json.loads(r.text)
    print(ur)
    
    if(y["examination"]=="GRADE 5 SCHOLARSHIP EXAMINATION"):
      print (r.text)
      collection3.insert_one( y)
      
```

```
#parallel program 
def parallel(targ,start, end):
  threadlist=[]
  t1=perf_counter()
  ur=start
  while ur < end:
    t = Thread(target=targ,args=(ur,))
    t.start()
    threadlist.append(t)
    ur=ur+1


  for b in threadlist:
    b.join()

  t2=perf_counter()-t1
  print("elapsed time :"+str(t2)+"s")
  print("requests made per second approx. :"+str((end-start)/t2))
```

```
parallel(alevel,8000000,8500000)
```

```
parallel(olevel,80000000,80000100)
```

```
parallel(scholarship,8000000,8250000)
```
